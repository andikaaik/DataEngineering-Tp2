### Melakukan plotting barplot menggunakan seaborn dengan estimator median. x = “Outcome” dan y = “BloodPressure”.

```python
data= pd.read_csv('diabetes.csv')
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Outcome', y='BloodPressure', data=data, estimator=np.median)

plt.xlabel('Outcome')
plt.ylabel('Blood Pressure')
plt.xticks(rotation=40)

plt.show()
```

Individu yang didiagnosis dengan diabetes memiliki tekanan darah rata-rata yang sedikit lebih tinggi.
Hal ini menunjukkan bahwa tekanan darah yang lebih tinggi mungkin lebih umum terjadi pada individu dengan diabetes, meskipun perbedaannya tidak terlalu ekstrem berdasarkan nilai rata-rata.

### melakukan plotting regplot menggunakan seaborn. X adalah semua kolom pada dataset diabetes, dan y = DiabetesPedigreeFunction

```python
y = data['DiabetesPedigreeFunction']

for col in data.columns:
    if col != 'DiabetesPedigreeFunction':
        plt.figure(figsize=(8, 6))
        sns.regplot(x=data[col], y=y)
        plt.title(f'Hubungan Antara {col} dan DiabetesPedigreeFunction')
        plt.xlabel(col)
        plt.ylabel('DiabetesPedigreeFunction')
        plt.show()
    else:
        continue
```

Secara umum, dari regplot-regplot tersebut dapat disimpulkan bahwa Glucose dan BMI memiliki hubungan yang lebih kuat dengan DiabetesPedigreeFunction dibandingkan variabel lainnya. Hubungan dengan variabel seperti Pregnancies, BloodPressure, dan SkinThickness tampak lebih lemah atau tidak signifikan.
